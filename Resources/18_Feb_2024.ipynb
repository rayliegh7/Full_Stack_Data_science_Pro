```python
#now we have to rename columns beacause most of columns names are not understand 
df2.rename(columns={'id':'id', 'age':'age', 'bp':'blood_pressure', 'sg':'specific_gravity', 'al':'albumin', 'su':'sugar', 'rbc':'red_blood_cells', 'pc':'pus_cell', 'pcc':'pus_cell_clumps', 'ba':'bacteria', 'bgr':'blood_glucose_random',
       'bu':'blood_urea', 'sc':'serum_creatinine', 'sod':'sodium', 'pot':'potassium', 'hemo':'hemoglobin', 'pcv':'packed_cell_volume', 'wc':'white_blood_cell_count', 'rc':'red_blood_cell_count', 'htn':'hypertension', 'dm':'diabetes_mellitus', 'cad':'coronary_artery_disease',
       'appet':'appetite', 'pe':'pedal_edema', 'ane':'anemia', 'classification':'class'},inplace=True)
```

```python
# here we are going to remove id cloumn
df2.drop('id',axis=1,inplace=True)
```

```python
df2.info()
```

```python
# find dubplicates records
df2.duplicated().sum()
```

```python
# now find numurical and object columns
object_columns = df2[[i for i in df2.columns if df2[i].dtypes=='object']]
```

```python
object_columns
```

```python
# here columns like packed_cell_volume,white_blood_cell_count,red_blood_cell_count they are actually numeric  but they are store as object 
# so first we have to convert their datatype to int or float
df2['packed_cell_volume'] = pd.to_numeric(df2['packed_cell_volume'],errors='coerce')
```

```python
# we can also done same thing using astype
# here we are unable to convert data type of white_blood_cell_count  and red_blood_cell_count to float beacuse it have some wrong entries like \t

# df2['red_blood_cell_count'] = df2['red_blood_cell_count'].astype('float64',errors='ignore')
try:
    df2['white_blood_cell_count'] = df2['white_blood_cell_count'].astype('float64',errors='raise')
except ValueError as ve:
    print(ve)
print(df2['white_blood_cell_count'].unique())
print(np.where(df2['red_blood_cell_count']=='\t?'))
print(np.where(df2['white_blood_cell_count']=='\t?'))
```

```python
df2[['red_blood_cell_count','white_blood_cell_count']].iloc[[185,162]]
```

```python
# we replace \t to nan and cthen used astype method 
# we can do these using str.replace method and replace

df2['red_blood_cell_count']=df2['red_blood_cell_count'].str.replace('\t?','nan')
```

```python
df2[['red_blood_cell_count','white_blood_cell_count']].iloc[[185,168]]
```

```python
df2['red_blood_cell_count'] = df2['red_blood_cell_count'].astype('float64',errors='raise')
print(df2['red_blood_cell_count'].dtype)
```

```python
df2['white_blood_cell_count']=pd.to_numeric(df2['white_blood_cell_count'],errors='coerce')
print(df2['white_blood_cell_count'].dtype)
```

```python
df2[['red_blood_cell_count','white_blood_cell_count']].iloc[[185,168]]
```

```python
object_columns_name = [col for col in df2.columns if df2[col].dtype=='object']
```

```python
df2[object_columns_name] # we have converted object columns to numeric data type using pd.to_numeric and .astype
```

```python
# now check unique value from each object columns
for col in object_columns_name:
    print(f'{col} : {df2[col].unique()}') # here  some wrong entries 'ckd\t' in column class diabetes_mellitus and coronary_artery_disease
```

```python
# we can solve these promblem using previous method
df2['class'] = df2['class'].str.replace('ckd\t','ckd')
print(df2['class'].unique())
```

```python
df2['coronary_artery_disease'] = df2['coronary_artery_disease'].replace(to_replace={'\tno':'no'})
print(df2['coronary_artery_disease'].unique())
```

```python
df2['diabetes_mellitus'] = df2['diabetes_mellitus'].replace(to_replace={'\tno':'no','\tyes':'yes',' yes':'yes'})
print(df2['diabetes_mellitus'].unique())
```

```python
# we have target labels class so lets find corelation with other labels
numeric_columns = df2[[col for col in df2.columns if df2[col].dtype!='object']]
```
