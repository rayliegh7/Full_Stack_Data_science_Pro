```python
# how to create series then convert series into df by changing index  
import numpy as np
data = np.random.rand(3,4).flatten() # beacuse series alway accept 1 d array 
sr = pd.Series(data) 
# now we to change indexing so we used paramter index
sr = pd.Series(data,index=[i for i in range(1,len(data)+1)]) # i have change index to int but from 1 to 12  prviously it was 0 to 11
sr = pd.Series(data ,index= ['a','b','c','d','e','f','g','h','i','j','k','l']) # here we have change into index to string
# Now we are going to convert series into data frame
df2 = pd.DataFrame(sr) 
# now we are going to change column namme
df2.columns =["ajay"]
```
