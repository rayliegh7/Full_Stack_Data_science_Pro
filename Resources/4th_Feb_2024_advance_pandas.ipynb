# accessing rows iloc & loc

Implexive Indexing: Refers to the default integer-based indexing provided by Pandas.
Explosive Indexing: Refers to custom indexing defined by the user.

2) Implexive Indexing:

This type of indexing uses integer positions to access data. For example, df.iloc is used for positional indexing.

The syntax is df.iloc[start_row:end_row, start_column:end_column].
Note that end_row and end_column are exclusive (i.e., they do not include the specified index).

3)Explosive Indexing:

This type of indexing allows for manual specification of index labels. For example, df.loc is used for label-based indexing.



The syntax is df.loc[start_row:end_row, start_column:end_column].
Note that end_row and end_column are inclusive (i.e., they  include the specified index).



| Feature                  | **`pd.merge()`**                                      | **`DataFrame.join()`**                                   |
|--------------------------|-------------------------------------------------------|----------------------------------------------------------|
| **Primary Use**           | Combines DataFrames using columns as keys.            | Combines DataFrames using their indexes (default).        |
| **Key or Index**          | Works with **column keys**. Can also use indexes with `left_index=True`, `right_index=True`. | Primarily works with **indexes**. Can use columns via `on`. |
| **Flexibility**           | More flexible: merges on columns or indexes, handles different column names. | Simpler: index-based joins mainly, limited column-based joins. |
| **Method**                | Standalone function: `pd.merge(df1, df2, ...)`.       | DataFrame method: `df1.join(df2, ...)`.                   |
| **Default Join Type**     | `inner` join (only matching rows).                    | `left` join (all rows from left DataFrame).               |
| **Column Duplication**    | Adds suffixes (`_x`, `_y`) for matching columns.      | Can lead to duplicate columns if same column names exist. |
| **Performance**           | More computationally intensive for complex merges.    | Faster for simpler index-based joins.                     |
| **Use Cases**             | Best for column-based joins, handling complex operations. | Best for simple index-based joins.                        |
| **SQL-style Joins**       | Supports `inner`, `outer`, `left`, `right` joins.     | Primarily `left` join, but supports other types (`inner`, `outer`, `right`). |
