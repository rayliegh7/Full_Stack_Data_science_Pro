```python
# here are some wrong entries in columns are  class,coronary_artery_disease,diabetes_mellitus we can change their wrong value using 
#  str.replace() and replace

df ['diabetes_mellitus']=df['diabetes_mellitus'].replace(to_replace={'\tno': 'no',' yes':'yes','\tyes':'yes'})
df['class'] = df['class'].str.replace('ckd\t','ckd')
df['coronary_artery_disease'] = df['coronary_artery_disease'].replace(to_replace=({'\tno':'no'}))
```

```python
for i in categorical_columns:
    print(f'{i} : {df[i].unique()}')
```

```python
df['class']=df['class'].apply(lambda x: 0 if x=='ckd' else 1)
```

```python
plt.figure(figsize=(20,15))
plotnumber=1
for i in numurical_columns:  # Ensure 'numerical_columns' is defined in your code
    if plotnumber <= 15:  # Ensure no more than 15 plots
        ax = plt.subplot(3, 5, plotnumber)  # 3 rows, 5 columns grid
        sns.histplot(data=df, x=i,kde=True)  # KDE plot using seaborn
        plt.title(i)  # Set the title as the column name
        plotnumber += 1

plt.tight_layout()
plt.show()
```

```python
def kde_plot(col,data):
    grid=sns.FacetGrid(data,hue='class',height=3,aspect=2)
    grid.map(sns.kdeplot,col)
    grid.add_legend()
```

```python
for i in numurical_columns:
    kde_plot(i,df)
```
